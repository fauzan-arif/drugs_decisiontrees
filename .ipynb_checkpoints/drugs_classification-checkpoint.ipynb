### Importing relevant packages 

```python
# Importing the required packages 
import numpy as np 
import pandas as pd 
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt 
import seaborn as sns
# Suppress warnings 
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df_drugs = pd.read_csv('drug200.csv')
```

```python
# We have 4 columns categorical columns and we will need to encode them, several ways to do it i.e one hot encoder, or label encoder
```

```python
#initialize the encoder class
label_enc = LabelEncoder()
```

```python
#iterating through each categorical column and encode it. 
cols_to_encode = ['Sex', 'BP', 'Cholesterol', 'Drug']
for col in cols_to_encode:
    df_drugs[col] = label_enc.fit_transform(df_drugs[col])
df_drugs
```

### Modelling and Model Evaluation

```python
#Lets seperate the features and target variable 
X = df_drugs.drop(columns = 'Drug')
Y = df_drugs['Drug']
```

```python
#Split the data intro train , test split 
X_train, X_test, y_train, y_test = train_test_split(  
    X, Y, test_size = 0.3, random_state = 100) 
```

```python
#initialize the decision tree using both entropy and gini metrics 
clf_gini = DecisionTreeClassifier(criterion = "gini")
clf_entropy = DecisionTreeClassifier(criterion = "entropy")
```

```python
#fitting the model using gini metric on features and target
clf_gini.fit(X_train,y_train)
```

```python
#fitting the model using entropy on features and target
clf_entropy.fit(X_train,y_train)
```

```python
#predicting the target and then evaluating the performance using classification repport
y_pred_gini = clf_gini.predict(X_test)
report_gini = classification_report(y_test, y_pred_gini)
print(report_gini)
```

```python
##predicting the target and then evaluating the performance using classification repport
y_pred_entropy = clf_entropy.predict(X_test)
report_entropy = classification_report(y_test, y_pred_entropy)
print(report_entropy)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 8), dpi=200)
plot_tree(clf_gini, feature_names=df_drugs.columns, filled=True)
```

```python
plot_tree(clf_entropy, feature_names=df_drugs.columns, filled=True)
```
